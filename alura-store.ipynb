# 3. Média de Avaliação das Lojas


```python
# Seleciona a coluna 'Avaliação da compra' do DataFrame da loja
# e calcula a média das avaliações feitas pelos clientes
media_avaliacao_loja = loja['Avaliação da compra'].mean()
media_avaliacao_loja
```

```python
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja2
```

```python
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja3
```

```python
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()
media_avaliacao_loja4
```

```python
# Lista com os nomes das lojas para facilitar a leitura dos resultados
nomes_lojas = ['Loja', 'Loja 2', 'Loja 3', 'Loja 4']

# Lista com as médias de avaliação de cada loja
medias_avaliacao = [
    media_avaliacao_loja,
    media_avaliacao_loja2,
    media_avaliacao_loja3,
    media_avaliacao_loja4
]

# Percorre as duas listas ao mesmo tempo e exibe o resultado formatado
for nome, media in zip(nomes_lojas, medias_avaliacao):
    print(f"{nome} - Média de avaliação: {media:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
# Conta quantas vezes cada produto aparece no DataFrame da loja
# Isso representa a quantidade de vendas por produto
contagem_produtos_loja = loja['Produto'].value_counts()
contagem_produtos_loja
```

```python
# Identifica o produto mais vendido da loja
produto_mais_vendido_loja = contagem_produtos_loja.idxmax()

# Identifica a quantidade de vendas do produto mais vendido
quantidade_mais_vendido_loja = contagem_produtos_loja.max()

# Identifica o produto menos vendido da loja
produto_menos_vendido_loja = contagem_produtos_loja.idxmin()

# Identifica a quantidade de vendas do produto menos vendido
quantidade_menos_vendido_loja = contagem_produtos_loja.min()

print(f"Loja - Produto mais vendido: {produto_mais_vendido_loja} ({quantidade_mais_vendido_loja} vendas)")
print(f"Loja - Produto menos vendido: {produto_menos_vendido_loja} ({quantidade_menos_vendido_loja} vendas)")

```

```python
contagem_produtos_loja2 = loja2['Produto'].value_counts()

produto_mais_vendido_loja2 = contagem_produtos_loja2.idxmax()
quantidade_mais_vendido_loja2 = contagem_produtos_loja2.max()

produto_menos_vendido_loja2 = contagem_produtos_loja2.idxmin()
quantidade_menos_vendido_loja2 = contagem_produtos_loja2.min()

print(f"Loja 2 - Produto mais vendido: {produto_mais_vendido_loja2} ({quantidade_mais_vendido_loja2} vendas)")
print(f"Loja 2 - Produto menos vendido: {produto_menos_vendido_loja2} ({quantidade_menos_vendido_loja2} vendas)")
```

# 5. Frete Médio por Loja

```python
# Seleciona a coluna 'Frete' do DataFrame da loja
# e calcula o valor médio pago pelos clientes em frete
frete_medio_loja = loja['Frete'].mean()
frete_medio_loja
```

```python
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja2
```

```python
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja3
```

```python
frete_medio_loja4 = loja4['Frete'].mean()
frete_medio_loja4
```

```python
# Lista com os nomes das lojas
nomes_lojas = ['Loja', 'Loja 2', 'Loja 3', 'Loja 4']

# Lista com os valores médios de frete de cada loja
fretes_medios = [
    frete_medio_loja,
    frete_medio_loja2,
    frete_medio_loja3,
    frete_medio_loja4
]

# Exibe o frete médio de cada loja de forma organizada
for nome, frete in zip(nomes_lojas, fretes_medios):
    print(f"{nome} - Frete médio: R$ {frete:.2f}")
```
